```python
import json
import pandas as pd
import copy

import geojson
from geojson import Feature, Point, FeatureCollection
```

```python
fn = "metro-bike-2020-q2.csv"
pth_quarter = "./DSCI-554/Project/project-love554/data/processed_data/" + fn
df_trip = pd.read_csv(pth_quarter)

l_trip = df_trip.to_dict('records')
l_trip[0]
```

```python
pth_station = "./DSCI-554/Project/project-love554/data/processed_data/metro-bike-stations.csv"
df_station = pd.read_csv(pth_station)
l_origin = df_station.to_dict('records')
l_origin[0]
```

```python
def getFeature(station):
    # (lon, lat)
    return Feature(id=station['station'], geometry=Point((station['lon'], station['lan'])), properties=station)
```

```python
fc = FeatureCollection([getFeature(s) for s in l_origin])
```

```python
fc
```

```python
list_station = copy.deepcopy(l_origin)
list_station[0]
```

```python
quarters = [
    '2016-q3',
    '2016-q4',
    '2017-q1',
    '2017-q2',
    '2017-q3',
    '2017-q4',
    '2018-q1',
    '2018-q2',
    '2018-q3',
    '2018-q4',
    '2019-q1',
    '2019-q2',
    '2019-q3',
    '2019-q4',
    '2020-q1',
    '2020-q2'
]
```

```python
to_add_per_quarter = {}
for stt in list_station:
    to_add_per_quarter[stt['station']] = {}
    for i in quarters:
        to_add_per_quarter[stt['station']][i] = {'tot_pop': 0, 'top_10': False}
to_add_per_quarter
```

```python
for i in quarters:
    fn = 'metro-bike-%s.csv' % i
    pth_quarter = "./DSCI-554/Project/project-love554/data/processed_data/" + fn
    df_trip = pd.read_csv(pth_quarter)
    l_trip = df_trip.to_dict('records')
    
    for stt in list_station:
        for t in l_trip:
            if t['start_station'] == stt['station']:
                to_add_per_quarter[stt['station']][i]['tot_pop'] += 1
            elif t['end_station'] == stt['station']:
                to_add_per_quarter[stt['station']][i]['tot_pop'] += 1
to_add_per_quarter
```

```python
temp = "./DSCI-554/Project/project-love554/demo/public/station/top_10/2017-q3.json"
df_trip1 = pd.read_json(temp)
l_trip1 = df_trip1.to_dict('records')
l_trip1
```

```python
for i in quarters:
    fn = '%s.json' % i
    pth_quarter = "./DSCI-554/Project/project-love554/demo/public/station/top_10/" + fn
    df_trip = pd.read_json(pth_quarter)
    l_trip = df_trip.to_dict('records')
    
    res = [sub['name'] for sub in l_trip]
    
    for j in res:
        to_add_per_quarter.get(j)[i]['top_10'] = True
```

```python
to_add_per_quarter
```

```python
for i in list_station:
    del i['go_live_date']
    del i['region']
    del i['status']
    i.update(to_add_per_quarter.get(i['station']))
```
