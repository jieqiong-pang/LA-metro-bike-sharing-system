```python
rs_df = stt_df[['region', 'station_name']]
```

```python
rs_df.head(5)
```

```python
for area in rs_df.region.unique():
    print(area)
```

```python
for tup in zip(stt_df.region, stt_df.station_name):
    print(tup)
```

```python
tree = {
    'name': 'LA Metro Bike',
    'branchset': []
}

for area in rs_df.region.unique():
    tree['branchset'].append({
        'name': area,
        'branchset': []
    })

for tup in zip(stt_df.region, stt_df.station_name):
    for area_branch in tree['branchset']:
        if area_branch['name'] == tup[0]:
            area_branch['branchset'].append({'name': tup[1]})
```

```python
tree
```

```python
with open('./chart_data/tree_of_station/tree_of_station.json', 'w', encoding='utf-8') as fp:
    json.dump(tree, fp)
```
