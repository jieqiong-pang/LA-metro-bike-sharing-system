```python
import pandas as pd
import numpy as np
import os
```

```python
dir_name = 'processed_data'
```

```python
file_list = os.listdir(dir_name)
```

```python
file_list.sort()
```

```python
file_list
```

```python
def getStationData(fp):
    df_location_collection = pd.read_csv(fp)
    df_start_station = df_location_collection[["start_station", "start_lat", "start_lon"]]
    df_start_station.columns = ['station', 'lat', 'lon']
    df_end_station = df_location_collection[["end_station", "end_lat", "end_lon"]]
    df_end_station.columns = ['station', 'lat', 'lon']
    pd_replicate_station = pd.concat([df_start_station, df_end_station])
    return pd_replicate_station
```

```python
stations = {}
for f in file_list:
    if f[11:13] == "20":
        print('processing', f)
        file_pth = os.path.join(dir_name, f)
        station_df = getStationData(file_pth)
        
        for r in station_df.iterrows():
            stations[int(r[1]['station'])] = [r[1]['lat'], r[1]['lon']]
```

```python
len(stations)
```

```python
stations_l = []
for sid in stations.keys():
    stations_l.append([sid, stations[sid][0], stations[sid][1]])
```

```python
stations_l
```

```python
df = pd.read_csv("rawdata/metro-bike-stations.csv", na_values="N/A")
```

```python
df_index_isnan = df[df["Region"].isnull()].index
```

```python
df_no_nan = df.drop(df_index_isnan)
```

```python
df_no_nan
```

```python
df_st = pd.DataFrame(columns=['Station_ID', 'lan', 'lon'], data=stations_l)
```

```python
df_st
```

```python
df_final = df_no_nan.merge(df_st, how="inner")
```

```python
df_final.columns = ['station', 'station_name', 'go_live_date', 'region', 'status', 'lat', 'lon']
```

```python
output_filepath = os.path.join(dir_name, "metro-bike-stations.csv")
```

```python
df_final.to_csv(output_filepath, index=False)
```
